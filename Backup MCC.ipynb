```python
import pandas as pd
import sqlite3

# Caminho do arquivo .db
db_path = 'database.db'

# Conectar ao banco de dados
conn = sqlite3.connect(db_path)

# Listar todas as tabelas disponíveis no banco de dados
query = "SELECT name FROM sqlite_master WHERE type='table';"
tables = pd.read_sql_query(query, conn)

print("Tabelas disponíveis no banco de dados:")
print(tables)
```

```python
# Dicionário para armazenar os DataFrames
dataframes = {}

# Iterar sobre todas as tabelas e carregar seus dados
for table in tables['name']:
    dataframes[table] = pd.read_sql_query(f"SELECT * FROM {table};", conn)
    print(f"Dados da tabela '{table}':")
    print(dataframes[table].head())
    print("\n")
```

```python
# Fechar a conexão com o banco de dados
conn.close()
```

```python
for table, df in dataframes.items():
    print(f"Tabela: {table}")
    print(df.head())
```

```python
df_combined = pd.concat(dataframes.values(), ignore_index=True)
print(df_combined.head())
```
